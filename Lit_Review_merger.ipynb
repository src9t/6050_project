```python
import pandas as pd
import numpy as np
```

```python
#read in and take a look
train = pd.read_csv('train.csv')
train
```

```python
##read in and take a look
test = pd.read_csv('test.csv')
test
```

```python
# the labels for the test set
submit = pd.read_csv('submit.csv')
submit
```

```python
submit.label.value_counts()
```

```python
# test merge and validate they merged properly
test_merge= pd.merge(test, submit,
                    how = 'outer',
                    on = 'id',
                    indicator = 'matched',
                    validate ='one_to_one')

#everything matched
test_merge.query("matched != 'both'")
#test_merge         
```

```python
# real merge this time
test_with_label= pd.merge(test, submit,
                    how = 'left',
                    on = 'id')
```

```python
#check to see what we got
test_with_label
```

```python
test_with_label.label.value_counts()
```
